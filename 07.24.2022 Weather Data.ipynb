```python
df_normals=df2[df2['datatype']=='MLY-TMAX-NORMAL']
df_normals.loc[:,'value']=df_normals.loc[:,'value']/10
# df_normals['F']=df_normals['value']*9/5+32
df_normals.loc[:,'date']=pd.to_datetime(df2.loc[:,'date']).astype('datetime64').dt.date
df_normals['M']=pd.DatetimeIndex(df_normals['date']).month
df_normals = df_normals[['M','value']]

df_normals[0:25]
```

```python
df_temp=df[df['datatype']=='TMAX']
df_temp['value']=df_temp['value']/10
df_temp['F']=df_temp['value']*9/5+32
df_temp['date']=pd.to_datetime(df_temp['date']).astype('datetime64').dt.date
df_temp['Y']=pd.DatetimeIndex(df_temp['date']).year
df_temp['M']=pd.DatetimeIndex(df_temp['date']).month
df_temp = pd.merge(df_temp, df_normals, on='M', how='inner')

df_temp[0:25]
```
